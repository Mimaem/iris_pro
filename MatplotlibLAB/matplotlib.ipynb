```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
x = np.array([2018,2019,2020,2021,2022,2023])
y = np.array([3000,2500,4200,600,1300, 1500])
plt.plot(x,y)
plt.show()
```

```python
df = pd.read_csv('Dataset/gdp.csv')
plt.plot(df['year'], df['GDP'], marker = '.', markersize = 10, mfc = 'r', linestyle = 'dashdot') 
## linestyle: dotted, dash, dashdot, solid, none
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Laos GDP')

plt.show()
```

```python
df = pd.read_csv('Dataset/gdp.csv')
plt.bar(df['year'], df['GDP'], color = ['r', 'c', 'b', 'g', 'y', 'b']) 
## linestyle: dotted, dash, dashdot, solid, none
plt.xlabel('Year', c = 'b')
plt.ylabel('GDP', c = 'w')
plt.title('Laos GDP')

plt.show()
```

```python
df = pd.read_csv('Dataset/gdp.csv')
plt.scatter(df['year'], df['GDP'], marker = '*', color = 'r', s = 50) 
## linestyle: dotted, dash, dashdot, solid, none
plt.xlabel('Year', c = 'b')
plt.ylabel('GDP', c = 'w')
plt.title('Laos GDP')

plt.show()
```

```python
df = pd.read_csv('Dataset/gdp.csv')
plt.pie(df['GDP'], labels= df['year'], autopct= '%.2f%%') 
## linestyle: dotted, dash, dashdot, solid, none
plt.xlabel('Year', c = 'b')
plt.ylabel('GDP', c = 'w')
plt.title('Laos GDP')

plt.show()
```
